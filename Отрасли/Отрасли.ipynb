```python
import pandas as pd

file_path = 'Объемы промышленного производства по видам экономической деятельности по РК (1990-2023).xlsx'
excel_data = pd.ExcelFile(file_path)
df = pd.read_excel(file_path, sheet_name=excel_data.sheet_names[0], skiprows=4)

df['Тип'] = df['Тип'].str.strip()

df = df.fillna('0')

for date in df.columns[1:]:
    df[date] = df[date].astype(str)  
    df[date] = df[date].str.replace('...', '0')  

for date in df.columns[1:]:
    df[date] = df[date].str.replace(',', '') 
    df[date] = df[date].astype(float)
    df[date] = df[date] * 1000 

df
```

```python
import pandas as pd
import plotly.express as px

# Фильтрация строк, где значение в столбце 2024 больше 1 триллиона и меньше 3 триллионов
df_filtered = df[(df[2023] > 3_000_000_000_000)]


region_data = df_filtered

# Переходим от широкой таблицы к длинной
region_data = region_data.melt(id_vars=['Тип'], var_name='Год', value_name='Значение')

# Преобразуем год в строковый формат, если он не является строкой
region_data['Год'] = region_data['Год'].astype(str)

# Преобразуем год в числовой формат, извлекая только 4 цифры
region_data['Год'] = region_data['Год'].str.extract('(\d{4})').astype(int)

# Строим график с использованием plotly.express
fig = px.line(region_data, x='Год', y='Значение', color='Тип', title='Производство по отрослям в тенге, > 3Т', markers=True)

# Показываем график
fig.show()
```

```python
# Список интересующих регионов
regions_of_interest = ["Переработка и консервирование мяса и производство мясной продукции",
"Переработка и консервирование рыбы, ракообразных и моллюсков",
"Переработка и консервирование фруктов и овощей",
"Производство растительных и животных масел и жиров",
"Производство молочных продуктов",
"Производство мукомольно-крупяных продуктов, крахмалов и крахмальных продуктов",
"Производство хлебобулочных, макаронных и мучных кондитерских изделий",
"Производство прочих продуктов питания",
"Производство готовых кормов для животных"]

# Выбираем данные для всех интересующих регионов
region_data = df[df['Тип'].isin(regions_of_interest)]

# Переходим от широкой таблицы к длинной
region_data = region_data.melt(id_vars=['Тип'], var_name='Год', value_name='Значение')

# Преобразуем год в строковый формат, если он не является строкой
region_data['Год'] = region_data['Год'].astype(str)

# Преобразуем год в числовой формат, извлекая только 4 цифры
region_data['Год'] = region_data['Год'].str.extract('(\d{4})').astype(int)

# Строим график с использованием plotly.express
fig = px.line(region_data, x='Год', y='Значение', color='Тип', title='Производство по пищевым отрослям в тенге', markers=True)

# Показываем график
fig.show()
```
